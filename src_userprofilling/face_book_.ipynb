```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import joblib

# Load the dataset
data = pd.read_csv(r"C:\Users\ABISHEKRS\Downloads\pseudo_facebook.csv\pseudo_facebook.csv")
data.columns = data.columns.str.strip()

# Check if 'userid' exists, drop it if present
if 'userid' in data.columns:
    data.drop(columns=['userid'], inplace=True)

# Handle missing values (if any) in 'gender' column
data['gender'].fillna('Unknown', inplace=True)

# Example: Fitting the LabelEncoder
label_encoder = LabelEncoder()
data['gender'] = label_encoder.fit_transform(data['gender'])  # Fit on training data

# Save the fitted LabelEncoder
joblib.dump(label_encoder, 'label_encoder.pkl')
print("Label Encoder saved successfully as 'label_encoder.pkl'.")
```
